# 10.파일 읽고 쓰기

- 텍스트 파일을 읽고 쓰는 기본적인 방법을 학습합니다.
- 좋은 글을 일어와 멋진 워드 클라우드를 만들어봅니다
- 파일을 다루는 상세한 기술이 필요한 경우 관련 자료를 찾아 학습하기를 권고합니다.

## 10.1.경로 확인

- 사용자마다 파일을 저장하고 읽는 기본 경로를 ‘홈 디렉토리‘ 라고 합니다.
- pathlib 패키지의 Path.home() 함수로 사용자 홈 디렉토리를 확인합니다.

```python
# 홈 디렉토리 확인
from pathlib import Path

print(Path.home())
```

- pathlib 패키지의 Path.cwd() 함수로 현재 작업 디렉토리를 확인합니다.

```python
# 작업 디렉토리 확인
from pathlib import Path

print(Path.cwd())
```

- Path.cwd().glob()를 사용해 디렉토리 안의 내용을 확인합니다.

```python
# 디렉토리 내용 확인
from pathlib import Path

files = Path.cwd().glob('*.*')
for f in files:
    print(f)
    
```

## 10.2.기본적인 파일 읽고 쓰기

**1) 파일 쓰기**


- w 모드로 파일을 열어 write() 메소드로 파일에 기록합니다.
- 경로를 지정하지 않으면 현재 작업 디렉토리에 기록됩니다.
- 새로운 파일이 만들어지며, 같은 이름의 파일이 있으면 제거되고 새로 만들어집니다.

```python
# 파일 열기
f = open('MyFile.txt', 'w')

# 파일 쓰기
f.write('안녕하세요?\n')

# 파일 닫기
f.close()
```

- mkdir()를 사용해 현재 작업 디렉토리에 새로운 디렉토리를 만듭니다.

```python
# 디렉토리 만들기
Path('Files').mkdir(exist_ok=True)
```

- Files/MyFile.txt 처럼 경로를 지정해 특정 디렉토리 안에 파일을 만듭니다.
- Windows에서 경로 구분자로 '\\' 를 사용하는 것과 달리 '/' 를 사용함에 주의합니다.

```python
# 파일 열기
f = open('Files/MyFile.txt', 'w')

# 파일 쓰기
f.write('모두들 안녕하세요?\n')

# 파일 닫기
f.close()
```

**2) 파일 읽기**

- r 모드로 파일을 열어 read() 메소드로 파일을 읽을 수 있습니다.
- 파일이 없으면 오류가 발생합니다.

```python
# 파일 열기
f = open('MyFile.txt', 'r')

# 내용 읽기
print(f.read())

# 파일 닫기
f.close()
```

- 경로를 지정해 특정 디렉토리 안의 파일을 읽어옵니다.

```python
# 파일 열기
f = open('Files/MyFile.txt', 'r')

# 내용 읽기
print(f.read())

# 파일 닫기
f.close()
```

**3) 파일 내용 추가**

- a 모드로 파일을 열어 write() 메소드로 내용을 추가할 수 있습니다.

```python
# 파일 열기
f = open('MyFile.txt', 'a')

# 내용 추가
f.write('\n만나서 정말 정말 반갑습니다!')

# 파일 닫기
f.close()
```

- 추가된 내용을 확인합니다.

```python
# 파일 열기
f = open('MyFile.txt', 'r')
print(f.read())
f.close()
```

**4) 파일 있으면 오류 발생**

- x 모드로 파일을 열면 같은 파일이 있는 경우 오류를 발생시킵니다.
- 파일이 없으면 w 모드로 파일은 연 결과가 됩니다.

```python
# 파일 열기 
f = open('MyFile.txt', 'x')

# 내용 쓰기
f.write('만나서 반갑습니다!')

# 파일 닫기
f.close()
```

- 같은 이름의 파일이 있어 오류가 나는 것에 대한 핸들링을 합니다.

```python
# 오류 핸들링
try: 
    f = open('MyFile_test.txt', 'x')
    f.write('만나서 반갑습니다!')
    f.close()
except FileExistsError:     #  이미 파일이 있으면
    print('같은 이름의 파일이 있습니다.')
else:                       # 그렇지 않으면
    print('파일 쓰기 성공했습니다.')
finally:                    # 마무리
    print('수고하셨습니다.')
```

## 10.3.파일을 다루는 편리한 메소드

**1) writelines() 메소드**

- writelines() 메소드를 사용해 행 리스트를 파일에 기록할 수 있습니다.

```python
# 파일 여러 줄 쓰기
hello = ['1안녕하세요?\n', '만나서 반갑습니다!\n', '우리 사이좋게 잘 지내요.\n']
f = open('MyFile.txt', 'w')
f.writelines(hello)
f.close()
```

- 추가된 내용을 확인합니다.

```python
# 파일 읽기
f =  open('MyFile.txt', 'r') 
print(f.read())
f.close()
```

**2) readlines() 메소드**

- readlines() 메소드를 사용해 모든 행을 한 번에 읽을 수 있습니다.
- 결과는 리스트가 되므로, 반복문으로 한 행 씩 읽을 수 있습니다.

```python
# 파일 한 번에 읽기
f = open('MyFile.txt', 'r')
result = f.readlines()
f.close()

# 내용 확인
print(result)
```

```python
# 반복문으로 한 행씩 표시
for txt in result:
    print(txt, end ='')
```

**3) readline() 메소드**

- readline() 메소드를 사용해 행 단위로 읽을 수 있습니다.
- 결과는 문자열이 됩니다.

```python
# 파일 읽기
f = open('MyFile.txt', 'r')

# 내용 읽기
print(f.readline(), end='')
print(f.readline(), end='')
print(f.readline(), end='')

# 파일 닫기
f.close()
```

- 반복문을 이용해 모든 행을 읽어올 수 있습니다.

```python
# 파일 읽기
f = open('MyFile.txt', 'r')
result = f.readline()

# 반복문으로 한 행씩 읽어 표시
while result:
    print(result, end='')
    result = f.readline()

# 파일 닫기
f.close()
```

## 10.4.워드 클라우드 만들기

- 마틴루터킹 목사의 'I have a Dream' 연설문을 가지고 워드 클라우드를 만들어 봅니다.

<img src='https://cdn.jornaldebrasilia.com.br/wp-content/uploads/2020/06/martin-luther-king-1000x600.png' width=640/>

- 워드 클라우드를 만드는 순서는 다음과 같습니다.

1) 텍스트파일로 저장하기

    - 읽을 파일을 만듭니다.
    - 여기서는 수동으로 어디선가 가져오겠지만,
    - 나중에는 크롤링을 통해서 웹에서 가져올 수도 있습니다.
    
    
2) 텍스트 전처리

    - 파일을 읽어서 단어 단위로 쪼갭니다. .split()
    - 단어별 빈도수를 계산하여 저장합니다.(여기서는 딕셔너리에 저장)
    - 조사들은 제거합니다.
    
3) 워드 클라우드 그리기

    - 라이브러리를 불러옵니다.(아나콘다에는 기본적으로 설치되지 않으니 직접 설치해야 합니다)
    - 워드 클라우드를 그립니다.

**1) 텍스트파일로 저장하기**

- 수업을 위해 이미 저장해 놓았습니다.

**2) 텍스트 전처리**

- 우선 파일을 읽어서 단어 단위로 잘라 리스트로 저장합니다.

```python
# 파일 읽기
file = open('Dream.txt', 'r', encoding='UTF8')
text = file.read() 
file.close()

# 공백을 구분자로 하여 단어 단위로 자르기
wordList = text.split()
# print(wordList)
print(len(wordList))
```

- 단어별 빈도수를 계산하여 저장합니다(여기서는 딕셔너리에 저장).

```python
# 중복 단어 제거
worduniq = set(wordList)

# 딕셔너리 선언
wordCount = {}

# 단어별 개수 저장
for w in worduniq:
    wordCount[w] = wordList.count(w)
print(len(wordCount))
# 제외 대상 조사 
del_word = ['the','a','is','are', 'not','of','on','that','this','and','be','to', 'from']

# 제외하기
# worduniq -= set(del_word)
# print(len(worduniq))
for w in del_word:
    if w in wordCount :
        del wordCount[w]
print(wordCount)
    
```

**3) 워드 클라우드 그리기**

- 필요한 패키지를 설치합니다.

```python
# 패키지 설치합
!pip install wordcloud
```

- 드디어! 워드 클라우드를 그립니다.

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 워드 클라우드 그리기
wordcloud = WordCloud(width=1000, 
                      height=600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud)
plt.show()
```

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 파일 열시
file = open('let_it_be.txt', 'r', encoding ='UTF8')
text = file.read() 
file.close()

# 공백을 구분자로 하여 단어 단위로 자르기
wordList = text.split()
# print(wordList)

# 중복 단어 제거
worduniq = set(wordList)

# 딕셔너리 선언
wordCount = {}

# 단어별 개수 저장
for w in worduniq:
    wordCount[w] = wordList.count(w)

# 제외 대상 조사 
del_word = ['the','a','is','are', 'not','of','on','that','this','and','be','to', 'from']

# 제외 추가하기

for w in worduniq :
    if wordCount[w] < 2:  #빈도수가 2 미만인 글자 제거
        del_word.append(w)
# 제외하기

for w in del_word:
    if w in wordCount :
        del wordCount[w]
print(wordCount)
    
    
# 워드 클라우드 그리기
wordcloud = WordCloud(font_path ='C:/Windows/Fonts/JOKERMAN.TTF',
                      colormap = 'Blues',
                      width=1000, 
                      height=600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud)
plt.show()



```
