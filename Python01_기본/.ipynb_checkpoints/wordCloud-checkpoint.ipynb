```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 파일 열시
file = open('let_it_be.txt', 'r', encoding ='UTF8')
text = file.read() 
file.close()

# 공백을 구분자로 하여 단어 단위로 자르기
wordList = text.split()
# print(wordList)

# 중복 단어 제거
worduniq = set(wordList)

# 딕셔너리 선언
wordCount = {}

# 단어별 개수 저장
for w in worduniq:
    wordCount[w] = wordList.count(w)

# 제외 대상 조사 
del_word = ['the','a','is','are', 'not','of','on','that','this','and','be','to', 'from']

# 제외 추가하기

for w in worduniq :
    if wordCount[w] < 2:
        del_word.append(w)
# 제외하기

for w in del_word:
    if w in wordCount :
        del wordCount[w]
print(wordCount)
    
    
# 워드 클라우드 그리기
wordcloud = WordCloud(font_path ='C:/Windows/Fonts/JOKERMAN.TTF',
                      colormap = 'Blues',
                      width=1000, 
                      height=600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud)
plt.show()



```
