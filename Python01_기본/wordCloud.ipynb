```python
#  
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#  
file = open('let_it_be.txt', 'r', encoding ='UTF8')
text = file.read() 
file.close()

#      
wordList = text.split()
# print(wordList)

#   
worduniq = set(wordList)

#  
wordCount = {}

#   
for w in worduniq:
    wordCount[w] = wordList.count(w)

#    
del_word = ['the','a','is','are', 'not','of','on','that','this','and','be','to', 'from']

#  

for w in worduniq :
    if wordCount[w] < 2:
        del_word.append(w)
# 

for w in del_word:
    if w in wordCount :
        del wordCount[w]
print(wordCount)
    
    
#   
wordcloud = WordCloud(font_path ='C:/Windows/Fonts/JOKERMAN.TTF',
                      colormap = 'Blues',
                      width=1000, 
                      height=600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud)
plt.show()



```
