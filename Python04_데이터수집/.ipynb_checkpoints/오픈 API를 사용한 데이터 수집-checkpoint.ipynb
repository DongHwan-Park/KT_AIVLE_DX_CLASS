### 2.1.1.데이터 가져오기

```python
# 라이브러리 불러오기
import urllib.request 
import json 
import pandas as pd 
```

```python
# 인증키와 주소
key = '5467434c7a6b6973363963576a536e'
start = 1
end = 1000
use_dt = 20220807
url = f'http://openapi.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/{start}/{end}/{use_dt}'
#       http://openapi.seoul.go.kr:8088/(인증키)/xml(형식)/CardSubwayStatsNew/1(시작)/5(끝)/20151101(일자)

# 데이터 가져오기
response = urllib.request.urlopen(url) 
json_str = response.read().decode('utf-8')

# JSON  -- > Dictionary 로 변환
json_object = json.loads(json_str)

# Dictionary --> DataFrame 으로 변환
subway = pd.json_normalize(json_object['CardSubwayStatsNew']['row'])

# 확인
subway.head()
```

- 서울특별시 공공자전거 실시간 대여정보를 가져와 **bike** 데이터프레임을 만듭니다.
- 데이터셋은 대여소별 실시간 자전거 대여가능 건수, 거치율, 대여소 위치정보를 제공합니다.
- 호출시 시스템 부하로 한번에 최대 **1,000건**를 초과할수 없습니다.
- 우선 1,000개 행만 조회하세요.

```python
# 인증키와 주소
key = '50564b6c476b697338385644694251'
start = 1
end = 1000

url = f'http://openapi.seoul.go.kr:8088/{key}/json/bikeList/{start}/{end}/'


# 데이터 가져와 인코딩
response = urllib.request.urlopen(url) 
json_str = response.read().decode('utf-8')

# JSON  -- > Dictionary 로 변환
json_object = json.loads(json_str)

# Dictionary --> DataFrame 으로 변환
bike = pd.json_normalize(json_object['rentBikeStatus']['row'])

# 확인
bike.head()
```

# 실습 #2: 서울특별시 공공자전거 실시간 대여정보  가져오기 #2

- 위 실습 결과로 작성된 구문을 수정하여 모든 데이터를 가져와 **bike** 데이터프레임을 다시 만듭니다.
- while 문을 사용해 JSON 문자열을 연결 한 후 이후 과정을 진행합니다.

```python
# 인증키
key = '50564b6c476b697338385644694251'

loop = 0
result = []

# 반복해서 가져와 연결하기
while True:
    start = 1 + loop * 1000 # 1 , 1001 , 2002
    end = 1000 + loop * 1000
    url = f'http://openapi.seoul.go.kr:8088/{key}/json/bikeList/{start}/{end}/'

    # 데이터 가져와 인코딩
    response = urllib.request.urlopen(url) 
    json_str = response.read().decode('utf-8')
    json_object = json.loads(json_str)
    
    result += json_object['rentBikeStatus']['row']
    
    
    if len(result) % 1000 != 0:  #1000에 다다르면 break
        break
        
    loop += 1
    
# 데이터프레임으로 변환
bike = pd.json_normalize(result)

# 상위 데이터 확인
bike.head()
```

```python
# 하위 데이터 확인
bike.tail()
```

```python
# 라이브러리 불러오기
import urllib.request 
import json 
import pandas as pd 
import xmltodict

# 미리보기주소
# https://apis.data.go.kr/1220000/nationtrade/getNationtradeList?serviceKey=인증키&strtYymm=202101&endYymm=202112&cntyCd=US'

# 매개변수
key = 'BjHZZcbP1j%2BVrenVf6W3sbp0PhiHL4LqmwwX8qgUMMPrzw4updVb9kO7numWldmB5V5nb2SJ59%2FW2V1t%2F%2FLOxQ%3D%3D'
strtYymm = 202101
endYymm = 202112
# cntyCd = '' #국가별

# 주소 https --> http 변경 이유?
url = f'http://apis.data.go.kr/1220000/nationtrade/getNationtradeList?serviceKey={key}&strtYymm={strtYymm}&endYymm={endYymm}'
# url = f'https://apis.data.go.kr/1220000/nationtrade/getNationtradeList?serviceKey={key}&strtYymm={strtYymm}&endYymm={endYymm}'

# 데이터 가져오기
response = urllib.request.urlopen(url) 

# xml --> dict
xml_parse = xmltodict.parse(response)
xml_dict = json.loads(json.dumps(xml_parse))
result = xml_dict['response']['body']['items']['item']  # 데이터 부분 딕셔너리 키 확인 필요

# 데이터프레임으로 변환
im_export = pd.json_normalize(result)

# 확인
im_export
```

# 실습 #4: 코로나19 시도별 발생 현황 가져오기

- 시.도별 코로나19 신규확진자, 신규사망자등에 대한 자료를 가져와 **covid19** 데이터프레임을 만듭니다.
- 기간은 2022.01.01 ~ 2022.06.30으로 합니다.

```python
# 라이브러리 불러오기
import urllib.request 
import json 
import pandas as pd 
import xmltodict

# 미리보기주소
# https://apis.data.go.kr/1220000/nationtrade/getNationtradeList?serviceKey=인증키&strtYymm=202101&endYymm=202112&cntyCd=US'

# 매개변수
# key = 'BjHZZcbP1j%2BVrenVf6W3sbp0PhiHL4LqmwwX8qgUMMPrzw4updVb9kO7numWldmB5V5nb2SJ59%2FW2V1t%2F%2FLOxQ%3D%3D'
key = 'P2sLgP4VJMxMK7JTjYfuKKm1oP5jb6gqWyipV8JqMwrN%2BnW1s6VnkST8Z7ofpV7TEe1ZSS9lAZ9ES9cN%2BxYwGg%3D%3D' #장래's key
startCreateDt = 20200101
CreateDt = 20220731

# 주소 https --> http 변경 이유?

url = f'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson?serviceKey={key}&pageNo=1&numOfRows=10&startCreateDt={startCreateDt}&endCreateDt={CreateDt}'

# 데이터 가져오기
response = urllib.request.urlopen(url) 

# xml --> dict
xml_parse = xmltodict.parse(response)
xml_dict = json.loads(json.dumps(xml_parse))
result = xml_dict['response']['body']['items']['item']  # 데이터 부분 딕셔너리 키 확인 필요

# 데이터프레임으로 변환
covid19 = pd.json_normalize(result)

# 확인
covid19
```

**데이터 셋 설명**

- createDt: 등록일시분초
- deathCnt: 사망자 수
- defCnt: 확진자 수
- gubun: 시도명(한글)
- gubunCn: 시도명(중국어)
- gubunEn: 시도명(영어)
- incDec: 전일대비 증감 수
- localOccCnt: 지역발생 수
- overFlowCnt:해외 유입 수
- qurRate: 10만명당 발생률
- seq: 게시글 번호(국내 시도별 발생현황 고유값)
- stdDay: 기준일시
- updateDt: 수정일시분초

```python
# 날짜만 남기기
covid19['createDt'] = pd.to_datetime(covid19['createDt']).dt.date
```

```python

cols = ['deathCnt', 'defCnt', 'incDec', 'localOccCnt', 'overFlowCnt'] 
for col in cols: 
    covid19[col] = covid19[col].astype('int')
```

```python
# 전일 대비 금일 증가된확진자 
covid19_all = covid19.groupby(by ='createDt',as_index = False)[['incDec']].sum()

# 확인
covid19_all
```

```python
plt.plot(covid19_all['createDt'], covid19_all['incDec'])
plt.ticklabel_format(axis='y', style='plain')
plt.show()
```

- 서울시 확진자 증가수를 라인 그래프로 시각화합니다.

```python
# 서울시 확진자 현황
# 집계
tmp = covid19.groupby(by ='createDt', as_index = False).sum()
tmp = tmp.sort_values(by ='createDt', ascending = True)
tmp.reset_index(drop= True, inplace = True)

tmp.columns = ['날짜', '사망자수', '확진자수', '전일대비증감수', '지역발생수','해외유입수']
tmp
```

```python
plt.title('코로나 진행')
plt.xlabel('월')
plt.ylabel('인원수')
plt.plot( tmp['날짜'], tmp['확진자수'])
plt.ticklabel_format(axis='y',style='plain') # y값을 지수형태가아닌 원래 값이 보니게함
plt.xticks(rotation = 70)
plt.legend(['확진자수'])
plt.show()
```
