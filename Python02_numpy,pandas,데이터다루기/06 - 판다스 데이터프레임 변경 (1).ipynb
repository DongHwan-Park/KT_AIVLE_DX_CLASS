# 5.데이터프레임 변경 (1)

- 데이터 전처리를 위해 꼭 익혀두어야 할 내용입니다.

```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/tips(2).csv'
tip =  pd.read_csv(path)

# 확인
tip
```

## 5.1.열 이름 변경

- 기존 데이터프레임의 열 이름을 적절히 변경해야 할 경우가 있습니다.
- 또는 집계 결과를 가진 열 이름을 이해하기 쉽게 변경해야 할 경우도 있습니다.

**1) 일부 열 이름 변경**

- **rename()** 메소드를 사용해 변경 전후의 열 이름을 딕셔너리 형태로 나열하는 방법으로 변경합니다.
- **inplace=True** 옵션을 설정해야 변경 사항이 실제 반영이 됩니다.
- 다음과 같이 열 이름을 변경합니다.
    - total_bill_amount → total_bill
    - male_female → sex
    - smoke_yes_no → smoker
    - week_name → day
    - dinner_lunch → time

```python
# rename() 함수로 열 이름 변경
tip.rename(columns = {'total_bill_amount' :'total_bill',
                      'male_female' : 'sex',
                      'smoke_yes_no': 'smoker',
                      'week_name' : 'day',
                      'dinner_lunch': 'time'}, inplace=True)  # inplace 까먹지말기
# 확인
tip.head()
```

**2) 모든 열 이름 변경** (선호)

- 모든 열 이름을 변경할 때는 **columns** 속성을 변경합니다.
- 변경이 필요없는 열은 기존 이름을 부여해 변경합니다. 


```python
# 모든 열 이름 변경:  
# ['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']
print(tip.columns)
tip.columns = ['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']
# 확인
tip.head()
```

## 5.2.열 추가

- 새로운 열을 추가하여 기존 데이터에서 계산된 결괏값을 저장해야할 경우가 있습니다.
- total_bill에 tip을 더한 값을 갖는 final_amt 열을 추가합니다.

```python
# final_amt 열 추가: final_amt = total_bill + tip
tip.loc[:,'final_amt'] = tip['total_bill'] + tip['tip']

# 확인
tip.head()
```

- insert() 메소드를 사용하면 원하는 위치에 열을 추가할 수 있습니다.


```python
# tip 열 앞에 div_tb 열 추가: div_tb = total_bill / size
tip.insert(1, 'div_tb', tip['total_bill']/ tip['size']) #(컬럼위치,컬럼명,element)

# 확인
tip.head()
```

## 5.3.열 삭제

- 뭔가를 삭제할 때는 **항상 조심x100** 해야 합니다.

**1) 열 하나 삭제**

- **drop()** 메소드를 사용해 열을 삭제합니다.
- axis=0: 행 삭제(기본 값) 
- axis=1: 열 삭제
- **inplace=True** 옵션을 지정해야 실제로 반영이 됩니다.


```python
# 열 하나 삭제: final_amt
tip.drop('final_amt', axis = 1, inplace = True)

# 확인
tip.head()
```

**2) 여러 열 삭제**

- 삭제할 열을 리스트 형태로 전달해 한 번에 여러 열을 제거할 수 있습니다.

```python
# 열 두 개 삭제: div_tb, day
del_cols = ['div_tb', 'day']  ## ex) col_x 로 지울 컬럼이라고 쓰심 feat.갓장래
tip.drop(del_cols, axis = 1, inplace = True)

# 확인
tip.head()
```

## 5.4.유익한 기능  (어려우면서,알면 편함)

- 값을 변경하거나 결괏값을 갖는 열을 추가하는 유익한 기능들을 익혀두면 좋습니다.

**1) map() 메소드**

- **map()** 메소드를 사용하면 범주형 값을 다른 값으로 쉽게 변경 할 수 있습니다.
- 다음 구문은 Male, Female을 각각 숫자 1, 0으로 변경합니다.

```python
# Male -> 1, Female -> 0 ==> {'Male': 1, 'Female': 0}
tip['sex'] = tip['sex'].map({'Male': 1, 'Female': 0})

# 확인
tip.head()
# tip_sex.head()
```

**2) cut() 함수**

- **크기**를 기준으로 범위를 나누어 등급을 지정하고 싶을 때 **cut() 함수**를 사용하면 좋습니다.
- 범위 개수를 지정하면 자동으로 크기를 기준으로 나눕니다.

```python
#문자열  --> 리스트
list('EDCBA')
```

```python
# tip 크기로 5등분해서 등급(a ~ e) 구하기
tip['tip_grp'] = pd.cut( tip['tip']   , 5  , labels = list('EDCBA')    )  #tip['tip'] 기준으로 5개 EDCBA로 나눈다

# 확인
tip.head()
```

```python
# 범주값 확인
tip['tip_grp'].value_counts()
```

```python
# 큰 값에 할당된 라벨이 큰 값으로 인식됨 tip크면== tip_grp A
tip.sort_values(by = 'tip', ascending= False)
```

```python
# tip 크기로 5등분해서 등급(a ~ e) 구하기  !! E->A로하는이유 totalbill 이작은수부터 나열되어있으므로 E부터 A로 할당
tip['total_grp'] = pd.qcut( tip['total_bill']   , 5  , labels = list('EDCBA')    )  #tip['tip'] 기준으로 5개 EDCBA로 나눈다

# 확인
tip.head()
```

```python
tip['total_grp'].value_counts()
```

- 임의의 범위를 직접 지정할 수 있습니다.
- 우선 tip 열의 기초통계량을 살펴봅니다.

```python
# tip열 기초통계량
tip.describe()
```


- 다음 구문은 아래 범위에 따른 값을 갖는 tip_grp 열을 추가합니다.
  - 2.0 이하 → 'a'
  - 2.0 초과 2.9 이하 → 'b'
  - 2.9 초과 3.5625 이하 → 'c'
  - 3.5625 초과 10.0 이하 → 'd' 
- 음의 무한대는 -np.inf, 양의 무한대는 np.inf로 지정합니다.

```python
# 라이브러리 불러오기
import numpy as np

# 등급 구하기
my_bin = [-np.inf, 2.0, 2.9, 3.5625, 10.0] # -np.inf 마이너스 무한대/ 경계값은 왼쪾에 포함 
label = ['a', 'b', 'c', 'd']
tip['tip_grp'] = pd.cut(  tip['tip'] , bins = my_bin  , labels = label )

# 확인
tip['tip_grp'].value_counts()
```

[믄2] 아래 범위와 값을 참조해 'a' ~ 'd' 값을 갖는 bill_grp 열을 추가하세요.

   - q1 이하 → 'a'
   - q1 초과 q2 이하 → 'b'
   - q2 초과 q3 이하 → 'c'
   - q3 초과 → 'd'

```python
# 등급 구하기
my_bin = [-np.inf, q1, q2, q3, np.inf]
my_label = list('abcd')

tip['bill_grp'] = pd.cut( tip['total_bill'], bins = my_bin, labels = my_label)


# 확인
tip.head()
# tip.info()
```

[문3] bill_grp 별 total_bill, tip의 평균을 집계하세요.

```python
# 평균
avg_bill_grp = tip.groupby(by = 'bill_grp', as_index= False)[['total_bill','tip']].mean()

# 확인
avg_bill_grp
```
