```python
# 열 이름 변수로 선언
org_col_name = list(tip.columns)

#확인
org_col_name
```

```python
# 새로운 열 이름
new_col_name = ['지불 금액', '팁', '성별', '흡연여부', '요일', '시간', '인원수']

# 열 이름 변경
tip.columns = new_col_name

#확인
tip.head()
```

```python
#orginal columns 로 복구
```

```python
# '흡연여부' 열 제거
cols_x =['흡연여부']
tip.drop(cols_x,axis =1, inplace = True)

tip.head()
```

```python
# 조건 조회
tip_tf = tip.loc[  tip['요일'].isin(['Thur','Fri'])  ,   :   ]

tip_tf.head()
```

```python
# 인덱스 초기화

tip_tf.reset_index(drop = True, inplace = True)

tip_tf.head()
```

```python
# 
tip_tf.groupby(by='요일', as_index = False)[['팁']].mean()
```

```python
#열 추가(없는 열을 변경하는 방법으로 추가)
tip['총금액'] = tip['지불 금액'] + tip['팁']

tip.head()
```

```python
#총금액 열 제거
tip.drop('총금액', axis = 1, inplace=True)

#특정 위치에 열추가
tip.insert(2, '총금액', tip['지불 금액'] + tip['팁'])

# 확인
tip.head()
```

```python
# 정렬 False = 내림차순 True = 오름차순
tip.sort_values(by ='총금액', ascending = False)
```
