```python
# 인증키와 주소

key = '5603003178'


#---------------------------------------------------------------------------------
result = []
# pageNo = 1
# 반복해서 가져와 연결하기

year_list = [year for year in range(2018,2022)]

for year in year_list :
    pageNo = 1
    print('reset pageNo',year)
     
    url = f'http://data.ex.co.kr/openapi/specialAnal/trafficFlowByTime?key={key}&type=json&iStdYear={year}'

    # 데이터 가져와 인코딩
    response = urllib.request.urlopen(url) 
    json_str = response.read().decode('utf-8')
    json_object = json.loads(json_str)

    result += json_object['trafficFlowByTimeLists']


# 확인
data = pd.json_normalize(result)
print(data.head(2))
print(data.shape)
```

```python
data.tail()
```

```python
tmp = data.loc[(data['sphlDfttScopTypeNm']=='D-3') | (data['sphlDfttScopTypeNm']=='D-1')|(data['sphlDfttScopTypeNm']=='D-2') ,:]
tmp
```

```python
tmp = tmp[['stdYear','stdHour','sphlDfttNm','sphlDfttScopTypeNm','trfl']]
tmp.head()
```

```python
tmp.columns = ['연도', '시간', '연휴', '추석기준', '통행량']
tmp.head()
```

```python
tmp['연도'] = tmp['연도'].str.slice(0,4)
tmp['연도']=tmp['연도'].astype(int)
tmp.head()
```

```python
tmp.info()
```

```python
df_date = pd.read_csv('df_date.csv')
del_cols = ['요일','구분','Unnamed: 0']
df_date.drop(columns = del_cols, axis =1, inplace = True)

df_date.head()
```

```python
df_date.info()
```

```python
date_data = pd.merge(df_date, tmp, how = 'inner', on = ['연도','추석기준'])
date_data
```

```python
date_data = date_data.loc[(date_data['연휴']=='추석'),:]
date_data.head()
```

```python
date_data.info()
```

```python
date_data['날짜시간'] = pd.to_datetime(date_data['날짜'].astype(str) + '-' + date_data['시간'].astype(str), format= '%Y-%m-%d-%H')
date_data.head()
```

```python
date_data.columns
```

```python
date_data = date_data[['날짜시간', '연도','추석기준','통행량']]
date_data['통행량'] = date_data['통행량'].astype(int)
date_data.head()
```

```python
date_data.info()
```

```python
date_data= date_data.sort_values(by = '날짜시간', ascending=True)
date_data.reset_index(drop=True, inplace = True)
date_data.head()
```
