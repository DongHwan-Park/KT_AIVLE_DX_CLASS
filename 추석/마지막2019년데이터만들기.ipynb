```python
# 라이브러리
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as spst

# 한글폰트 지정(맑은고딕)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 고해상도
%config InlineBackend.figure_format = 'retina'
```

```python
df_date = pd.read_csv('df_date_data.csv')
df_date=df_date.drop('Unnamed: 0',axis =1)
df_date['날짜시간'] = pd.to_datetime(df_date['날짜시간'].str.slice(0,13) ,format = '%Y-%m-%d %H')
df_date['시간'] = df_date['날짜시간'].dt.hour
df_date
```

```python
df_date.info()
```

```python
df_time = pd.read_csv('df_data_chu.csv')
df_time = df_time.drop(columns = ['출발지','도착지','연휴길이','연휴'])
df_time
```

```python
df_time.info()
```

```python
data = pd.merge(df_time ,df_date, how = 'inner', on =['연도','시간','추석기준'])
data
```

```python
data.to_csv('임시.csv')
```

```python
data_2019 = data.loc[(data['연도']==2019),:]
data_2019.reset_index(drop=True, inplace = True)
data_2019= data_2019[['날짜시간','연도','시간','추석기준','통행량','소요시간(분)']]
data_2019
```
