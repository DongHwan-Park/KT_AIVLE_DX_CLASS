```python
# 인증키와 주소

# key = '5603003178'
# iYear = 2022
# numOfRows = 23
# pageNo = 1
#---------------------------------------------------------------------------------
result = []
pageNo = 1
# 반복해서 가져와 연결하기
while True:

    url = f'http://data.ex.co.kr/openapi/specialAnal/intercityLeadTime?key=5603003178&type=json&iYear=2022&iStUnitCode=101&iEdUnitCode=140&numOfRows=23&pageNo={pageNo}'

    # 데이터 가져와 인코딩
    response = urllib.request.urlopen(url) 
    json_str = response.read().decode('utf-8')
    json_object = json.loads(json_str)
    
    result += json_object['intercityLeadTimeLists']
    
    
    if pageNo ==  11:  # 11페이지가 마지막
        break
    pageNo += 1    
    
# 확인
dochak = pd.json_normalize(result)
dochak.head()
```
