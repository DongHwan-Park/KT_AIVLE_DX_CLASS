## 3.데이터 전처리

전처리 과정에서 다음과 같은 처리를 할 수 있습니다.

- 결측치 처리
- 값 변경
- 열 추가
- 불필요한 열 제거
- 열 이름 변경
- 데이터 통합(연결 또는 조인)
- 가변수화 등

**1) 결측치 제거**

- 운동량, 탄소량 값이 결측치가 있는 행이 동일하고 몇 건 안되니 해당행을 제거합니다.

```python
# 결측치 제거
bike.dropna(axis = 0, subset = ['운동량','탄소량'], inplace = True)

# 확인
bike.info()
```

**2) 결측치 채우기**

- 성별 열은 결측치가 너무 많습니다. 
- 성별별 분석 시 '남','녀' 외에 '기타'로 분석되게 'X'로 채웁니다.

```python
bike.head(30)
```

```python
# 결측치 채우기
bike['성별'].fillna('x', inplace = True)

# 확인
bike.info()
```

```python
bike.head(50)
```

**3) 데이터 변경**

- 성별이 소문자 'm'. 'f'인 값을 대문자 'M'. 'F'로 변경합니다.

```python
# # 성별 소문자 --> 대문자 변경
# # sex = [{'m': 'M',
# #        'f': 'f'}]
# bike['성별'].str.replace('m','M')
# # bike = bike['성별'].str.replace('f','F')
# # 확인
# bike['성별'].value_counts()

bike['성별'] = bike['성별'].str.upper()
bike['성별'].value_counts()
```

- 연령대 코드를 이해하기 쉽게 '10대'~'80대'로 변경합니다.

```python
# 연령대 데이터 변경
bike['연령대코드'] = bike['연령대코드'].str[-1] + '0대'

# 확인
bike.head()
```

**4) 불필요한 열 제거**

- 분석에 큰 의미가 없는 대여일자, 대여소번호 열은 제거합니다.

```python
# 대여일자, 대여소번호 제거
bike.columns
bike = bike[['대여소명', '대여구분코드', '성별', '연령대코드', '이용건수', '운동량', '탄소량',
       '이동거리(M)', '이용시간(분)']]

# 확인
bike.head()
```

**5) 열 이름 변경**

- 분석에 편의를 위해 열 이름을 간단명료하게 변경합니다.
    - 대여소명 --> 대여소
    - 대여구분코드 --> 대여구분
    - 연령대코드  --> 연령대
    - 이동거리(M) --> 이동거리
    - 이용시간(분) --> 이용시간

```python
# 열 이름 변경
re_cols = ['대여소','대여구분','연령대','이동시간','이용시간']
bike.rename(columns = {'대여소명': '대여소',
                       '대여구분코드': '대여구분',
                       '연령대코드': '연령대',
                       '이동거리(M)': '이동거리',
                       '이용시간(분)': '이용시간',
                      }, inplace = True)


# 확인
bike.head()
```

## 4.데이터 분석

- 데이터를 집계하고 간단히 시각화하여 분석합니다.
- 연속형 값은 선 그래프, 범주형 값은 막대 그래프로 시각화함이 일반적입니다.

**1) 대여구분별 이용 건수**

- 대여구분별 이용 건수를 막대 그래프로 시각화해 확인합니다.
- 이용건수 합 단위가 크니 1,000으로 나누어 시각화에 사용합니다.

```python
# 대여구분별 이용건수

bike_use = bike.groupby(by = '대여구분', as_index = False)['이용건수'].sum()


# 시각화
plt.title('대여구분 별 이용 건수', size = 20, pad = 15)
plt.bar(bike_use['대여구분'], bike_use['이용건수']/1000)

plt.xticks(rotation=70) # 축 값 회전
plt.xlabel('대여구분')
plt.ylabel('이용건수(천)')
plt.show()
```

**2) 대여소별 이용 건수 TOP 10**

- 대여소별 이용 건수 TOP 10을 막대 그래프로 시각화해 확인합니다.

```python
# 대여소별 이용건수 TOP 10

tmp = bike.groupby(by = '대여소', as_index = False)['이용건수'].sum()
tmp = tmp.sort_values(by ='이용건수', ascending= False).head(10)


# 시각화
plt.title('대여소 별 이용 건수', size = 20, pad = 15)
plt.bar(tmp['대여소'], tmp['이용건수'])

plt.xticks(rotation=70) # 축 값 회전
plt.xlabel('대여소')
plt.ylabel('이용건수')
plt.show()


```

**3) 연령대별 이용 건수**

- 연령대별 이용 건수를 막대 그래프로 시각화해 확인합니다.

```python
# 연령대별 이용건수
tmp = bike.groupby(by = '연령대', as_index = False)['이용건수'].sum()

# 시각화
plt.title('연령대별 이용 건수', size = 20, pad = 15)
plt.bar(tmp['연령대'], tmp['이용건수']/1000)

plt.xticks(rotation=70) # 축 값 회전
plt.xlabel('연령대')
plt.ylabel('이용건수(천)')
plt.show()
```

**4) 성별별 이용 건수**

- 성별별 이용 건수를 막대 그래프로 시각화해 확인합니다.

```python
# 성별별 이용건수

tmp = bike.groupby(by = '성별', as_index = False)['이용건수'].sum()


# 시각화
plt.title('대여소 별 이용 건수', size = 20, pad = 15)
plt.bar(tmp['성별'], tmp['이용건수']/1000)

plt.xlabel('성별')
plt.ylabel('이용건수(천)')
plt.show()



```

**5) 이용시간 TOP 25**

- 이용시간을 내림차순으로 정렬해 상위 25개를 시각화해 확인합니다.

```python
# 성별별 이용건수
tmp = bike.sort_values(by = '이용시간', ascending = False).head(25)
tmp.reset_index(drop = True,inplace= True)

#시각화  --> 인덱스 리셋을 이용해서 
plt.title('이용시간 Top 25', size = 20, pad = 15)
plt.bar( tmp.index,height = tmp['이용시간'])

plt.xlabel('')
plt.ylabel('이용시간')
plt.xticks(rotation=70)
plt.show()
```

**6) 이용시간 TOP 25**

- 평균이용시간 열을 추가하여 이 값을 기분으로 내림차순 정렬해 상위 25개를 시각화해 확인합니다.

```python
# 이용시간
bike['평균이용시간'] = bike['이용시간'] / bike['이용건수']
tmp = bike.sort_values(by = '평균이용시간', ascending = False).head(25)
tmp.reset_index(drop = True , inplace = True)

#시각화
plt.title('평균이용시간 Top 25', size = 20, pad = 15)
plt.bar( tmp.index ,height = tmp['평균이용시간'])

plt.xlabel('')
plt.ylabel('평균이용시간')
plt.xticks(rotation=70)
plt.show()

```
