## 1.<데이터 조인>

```python
#조인 # 결측치 없음
origin = pd.merge(products,orders, on = 'ProductID', how = 'inner' )
origin = pd.merge(origin, customers, on = 'CustomerID' , how = 'inner')
```

```python
origin.head()
```

```python
# 열 삭제  큰 틀에서 가장 의미없는 열 앞으로 안쓸열 삭제

del_cols = ['Address','SubCategory','CategoryOrd','Seq','OrderID']
tmp = origin.drop(del_cols , axis = 1 )

#열 이름 변경 
tmp.columns = ['ProductID', 'ProductName', 'Category', 'OrderDate', 'Qty', 'Amt',
       'CustomerID', 'RegisterDate', 'Gender', 'BirthYear', '시도', '구군']

# total_bill = Qty * Amt 추가
tmp['total_bill'] = tmp['Qty'] * tmp['Amt']

# OrderYear , OrderMonth 추가, Age 추가
tmp['OrderYear'] = pd.to_datetime(tmp['OrderDate']).dt.year
tmp['OrderMonth'] = pd.to_datetime(tmp['OrderDate']).dt.month 
tmp['Age'] = 2017 - tmp['BirthYear']

# 이어서 불필요해진 del_cols2 = ['OrderDate','BirthYear'] 추가적으로 삭제
del_cols2 = ['OrderDate','BirthYear']
tmp.drop(del_cols2, axis = 1 ,inplace = True )

tmp.head()
```

**최종 가공 데이터**
- tmp 를 test_df 으로 저장

```python
test_df = tmp.copy()

test_df.head()
```

```python
# test_df 의 info() / describe() /결측치 확인

print(test_df.info())
print('-' * 50)
print(test_df.describe().apply(lambda x : x.apply('{:,.1f}'.format)))
print('-' * 50)
test_df.isna().sum()

#결과 결측치 없고 68839 rows 로 origin 과 같은 행 크기로 잘 제작된 모습
```

## 2.<데이터 분석>

- 남녀별 판매 현황을 막대 그래프로 시각화해 확인합니다.

## 3. <고객분류>

- 우선 고객을 일정 기준에 따라 분류하고 등급을 지정해야 합니다.
- 다음과 같은 순서로 처리하면 될 것입니다.
    - 고객별 주문 수량 합, 판매 금액 합을 갖는 데이터프레임 선언
    - 고객 분류 기준과 이 기준에 따른 등급 이름 정의
    - 고객 등급을 갖는 새로운 Level 열 추가 ( Merge??? )

```python
test_df.head(2)
```

**1) 고객별 집계**

* 고객별 주문 수량 합, 판매 금액 합을 갖는 cst 데이터프레임을 선언합니다.

```python
# CustomerID 를 기준으로 수량,  총판매액 집계

cst = test_df.groupby(by ='CustomerID', as_index = False)[['Qty','total_bill']].sum()

cst.tail()
```

**2) 고객 분류 기준 찾기**

- 고객 등급 분류 기준을 찾기 위해 관련 데이터를 확인합니다.

```python
# 고객 등급 분류 기준을 위해 관련데이터 확인 describe() 이용하여 총매출의 25%,50%,75% 값 이용

cst.describe().apply(lambda x : x.apply('{:,.1f}'.format))
```

**3) 고객 분류 기준 정리**

- 고객 등급 분류 기준과 그 기준에 따른 등급 이름을 정리합니다.
- 예를 들면 다음과 같은 기준과 등급 이름을 정의할 수 있을 것입니다.
    - ~ 30,000: Family
    - 30,000 ~ 100,000: Bronze
    - 100,000 ~ 300,000: Silver
    - 300,000 ~ : Gold
    - 600,000 ~ : Platinum
- 분류 기준과 등급 이름
- '총판매액' 을 기준으로 함
    - descirbe() 를 통해 
    -  Falmily ~ 3000
    -  Bronze ~ 9,850 25프로
    -  Silver ~ 37,410 50프로
    -  Gole ~137,210 ,75%
    - Platinum  137,210~ 

```python
q1 = cst.describe().loc['25%','total_bill']
q2 = cst.describe().loc['50%','total_bill']
q3 = cst['total_bill'].mean()
q4 = cst['total_bill'].mean() + 250000
my_bin = [-np.inf, q1, q2, q3, q4, np.inf] # -np.inf 마이너스 무한대/ 경계값은 왼쪾에 포함 
label = ['Family', 'Bronze', 'Silver', 'Gold','Platinum']

cst['등급'] = pd.cut( cst['total_bill'] , bins = my_bin  , labels = label )
# 확인
cst['등급'].value_counts()

#타당해보임
```

```python
cst_test = cst[['CustomerID','등급']]
cst_test
```

```python
last_test_df = pd.merge(test_df,cst_test, on='CustomerID', how = 'inner')
last_test_df
```
